```python
def calc_metrics(cf_matrix):
  tn = cf_matrix[0]
  fp = cf_matrix[1]
  fn = cf_matrix[2]
  tp = cf_matrix[3]
  if (((tp+fn)==0)|((tn+fp)==0)|((tp+fp)==0)|((tn+fp)==0)|((tn+fn)==0)):
    sens = tp/(tp+fn)
    spec = tn/(tn+fp)
    prec = 0
    acc = (tp+tn)/(tp+tn+fn+fp)
    mcc = 0
    f1 = tp/(tp+(0.5*(fp+fn)))
    return tp,fp,fn,tn,sens,spec,prec,acc,mcc,f1
  else: 
    sens = tp/(tp+fn)
    spec = tn/(tn+fp)
    prec = tp/(tp+fp)
    acc = (tp+tn)/(tp+tn+fn+fp)
    mcc = ((tp*tn)-(fp*fn))/(((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn))**0.5)
    f1 = tp/(tp+(0.5*(fp+fn)))
    return tp,fp,fn,tn,sens,spec,prec,acc,mcc,f1
```
